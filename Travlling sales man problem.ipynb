```
import sys  # Library for INT_size
 
class Graph():
 
    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[0 for column in range(vertices)] 
                      for row in range(vertices)]
 
    # A utility function to print the constructed MST stored in parent[]
    def printMST(self, parent):
        #print ("Edge \t \tWeight")
        print ("Edge ")
        alist=['UMERKOT','MITHI','MIRPURKHAS','SANGHAR','NAWABSHAH','HYDRABAD','karachi']
        for i in range(1,self.V):
            #print (parent[i],"-",i,"\t")
            #print (parent[i],"-",i,"\t",self.graph[i][ parent[i] ])
            print(alist[parent[i]],"-",alist[i],"\t",self.graph[i][ parent[i] ])
 
    # A utility function to find the vertex with minimum distance value, from
    # the set of vertices not yet included in shortest path tree
    def minKey(self, key, mstSet):
 
        # Initilaize min value
        min = sys.maxsize
 
        for v in range(self.V):
            if key[v] < min and mstSet[v] == False:
                min = key[v]
                min_index = v
 
        return min_index
 
    # Function to construct and print MST for a graph represented using
    # adjacency matrix representation
    def primMST(self):
 
        #Key values used to pick minimum weight edge in cut
        key = [sys.maxsize] * self.V
        parent = [None] * self.V # LIST to store constructed MST
        key[0] = 0   # Make key 0 so that this vertex is picked as first vertex
        mstSet = [False] * self.V
 
        parent[0] = -1  # First node is always the root of
 
        for cout in range(self.V):
 
            # Pick the minimum distance vertex from the set of vertices not
            # yet processed. u is always equal to src in first iteration
            u = self.minKey(key, mstSet)
 
            # Put the minimum distance vertex in the shortest path tree
            mstSet[u] = True
 
            # Update dist value of the adjacent vertices of the picked vertex
            # only if the current distance is greater than new distance and
            # the vertex in not in the shotest path tree
            for v in range(self.V):
                # graph[u][v] is non zero only for adjacent vertices of m
                # mstSet[v] is false for vertices not yet included in MST
                # Update the key only if graph[u][v] is smaller than key[v]
                if self.graph[u][v] > 0 and mstSet[v] == False and key[v] > self.graph[u][v]:
                        key[v] = self.graph[u][v]
                        parent[v] = u
 
        self.printMST(parent)
  

g  = Graph(7)                                 #             UMERKOT MITHI MIRPURKHAS SANGHAR NAWABSHAH HYDRABAD KARACHI
g.graph = [  [0 ,100 , 78,0 , 0, 0, 350],     # UMERKOT       
             [100, 0, 160, 0, 0, 200, 0],     # MITHI           
             [78, 160, 0, 80, 0, 70, 0],      # MiRPURKHAS         
             [0, 0, 80, 0,85,120,0],          #sANGHAR 
             [0, 0, 0, 85, 0, 140, 0],        #NAWABSHAH
             [0, 200, 70, 120, 140, 0, 160],  #HYDRABAD
             [0, 350, 0, 0, 0, 160, 0]        #KARACHI
             
             
           ]

print("============ORIGINAL TREE===============")

print("UMERKOT --78--- MIRPURKHAS ---80---SANGHAR")
print("  |               /  |             /   |")
print("  |              /   |            /    |")
print("  |             /    |           /     |")
print("  |            /     |          /      |")
print(" 100       160      70         120     |")
print("  |        /         |        /        85 ")
print("  |       /          |       /         |")
print("  |      /           |      /          |")
print("  |     /            |     /           |")
print("  |    /             |    /            |")
print("  |   /              |   /             |")
print("MITHI ---200---- HYDRABAD ---140----NAWABSHAH")
print("     \              |                      ")
print("      \             |                      ")
print("       \            |                      ")
print("        \           |                      ")
print("         \          |                      ")
print("         350        |                      ")
print("           \       160                      ")
print("            \       |                      ")
print("             \      |                      ")
print("              \     |                      ")
print("               \    |                      ")
print("                \   |                      ")
print("                 \  |                      ")
print("                 KARACHI                   ")
   
g.primMST()


print("========== MINIMUM SPINNING TREE============ ")

print("UMERKOT ---78--- MIRPURKHAS---80---SANGHAR")
print("  |                |                |")
print("  |                |                |")
print("  |                |                |")
print("  |                |                |")
print(" 100              70                85")
print("  |                |                |")
print("  |                |                |")
print("  |                |                |")
print("  |                |                |")
print("  |                |                |")
print("  |                |                |")
print("MITHI          HYDRABAD         NAWABSHAH")
print("                   |                      ")
print("                   |                      ")
print("                   |                      ")
print("                   |                      ")
print("                   |                      ")
print("                  160                      ")
print("                   |                      ")
print("                   |                      ")
print("                   |                      ")
print("                   |                      ")
print("                   |                      ")
print("                   |                      ")
print("                   |                      ")
print("                 KARACHI                   ")



 
```
